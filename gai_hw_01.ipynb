畫函數圖形

畫 $f(x) = sin(x)$

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel("x (radians)")  # Updated x-axis label
plt.ylabel("sin(x)")
plt.title("Plot of sin(x)")
plt.grid(True)

# Set x-axis ticks and labels in radians
plt.xticks(
    [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
    ['0', 'π/2', 'π', '3π/2', '2π']
)

plt.show()
```
